## Q1: Find out which Gender visits the Mall

```python
# Checking the content of Gender
mc.Gender.unique()
```

```python
# Finding the total number of Male
df_male = mc['Gender'] == 'Male'
total_male = df_male.sum()

# Printing total Male
print("Total number of Male:", total_male)
```

```python
# Finding the total number of Female
df_female = mc['Gender'] == 'Female'
total_female = df_female.sum()

# Printing total Female
print("Total number of Female:", total_female)
```

```python
# Gender that visits most
if total_male < total_female:
    print("females visit the Mall more.")
else:
    print("Males visit the Mall more.")
```

```python
# visializing Gender with piechart
labels = ['Females', 'Male']
size = mc['Gender'].value_counts()

fig, ax = plt.subplots()

ax.pie(size, explode=[0,0.05], labels = labels, shadow=True)
plt.title('Gender', fontsize=30)
plt.legend(loc='upper right', frameon=False)
plt.show()
```

## Q2: Find out what age group has the higher Annual Income

```python
fig.savefig('mall_analysis/artifacts/Gender.pdf', format='pdf')
```

```python
fig.savefig('mall_analysis/images/Gender_pie.jpg')
```

```python
age = mc['Age']
income = mc['Annual_Income_(k$)']

plt.rcParams['figure.figsize'] = (12, 3)
plt.title('Age/Annual Income')

sns.boxenplot(x=age, y=income)
plt.savefig('mall_analysis/artifacts/Income.pdf', format='pdf')
plt.savefig('mall_analysis/images/Age_Income.jpg')
plt.show()
```

<p> From the plot above, we found out that the age group with higher annual income are 32 - 33 </p>

```python
# Ploting Age with Annual income using line plot

age = mc['Age']
income = mc['Annual_Income_(k$)']

plt.rcParams['figure.figsize'] = (12, 3)

sns.lineplot(x=age, y=income)
plt.savefig('mall_analysis/artifacts/lineplot.pdf', format='pdf')
plt.savefig('mall_analysis/images/Age_Lineplot.jpg')
plt.show()
```

## Q3: Find out what age group spends more

```python
age = mc['Age']
score = mc['Spending_Score']

plt.rcParams['figure.figsize'] = (12, 3)
plt.title('Age/Spending Score')

sns.boxenplot(x=age, y=score)
plt.savefig('mall_analysis/images/spending_score.jpg')
plt.show()
```

<p>We found people around 35 years spends more at the mall.</p>
